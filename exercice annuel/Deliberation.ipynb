#### fichier excel en data fram

```python
import pandas as pd
import numpy as np
An=pd.ExcelFile('Del_Annuel.xlsx')
An=An.parse('PV')
An
```

#### 2. Supprimer les Nan

```python
Df1=An.dropna()
Df1
```

#### Remplacer le nom des colonnes par la ligne 9.

```python
col=An.iloc[9,:]
col
```

```python
Df2=Df1.rename(columns=col)
Df2
```

#### 4. Remplacer 'RESULTAT' par 'TotalCredit' et 'NaN' par 'Decision'

```python
Df2.rename(columns={'RESULTAT':'Total_Credit'},inplace=True)
Df2.columns=Df2.columns.fillna('Decision')
Df2.reset_index(drop=True)
```

#### 5. Trie Total_credit par decroissant et nom et prenom ordre croissant.

```python
tri_Df=Df2.sort_values(by=['Total_Credit','Nom','Prénom'],ascending=[False,True,True])
```

#### 6. enregistrer le Dataframe.

```python
import pandas as pd
# Cette fonction créera le fichier.
writer = pd.ExcelWriter("Cas_pratique_Annuel1.xlsx", engine="xlsxwriter")
# Ecriture du DataFrame dans le fichier : tri_df
tri_Df.to_excel(writer, "Deliberation",index=False)#CREATION DUNE FEUILLE "Deliberation".
# Enregistrement du résultat.
writer.save()
```

#### 7. Donner le pourcentage

```python
A=tri_Df.Decision.value_counts()
B=A.index
P=[]
for i in range(len(A)):
    if i==1:
        r=round(A[i]*100/len(tri_Df),1)+0.1
    else:
        r=round(A[i]*100/len(tri_Df),1)
    print("Le nombre d'étudiants en",B[i],"est de:",r)
    P.append(r)
```

```python
B=list(B)
B.append('Effectif_Total')
ar=np.array([[A[0],P[0]],[A[1],P[1]],[A[2],P[2]],[A[3],P[3]],[len(tri_Df),100]])
recap=pd.DataFrame(ar,index=B,columns=['Effectif','Pourcentage'])
recap
```

```python
import pandas as pd
# Cette fonction créera le fichier.
writer1 = pd.ExcelWriter("Cas_pratique_Annuel1.xlsx", engine="openpyxl",mode='a')
# Ecriture du DataFrame dans le fichier : tri_df
recap.to_excel(writer1, "Statistique")#CREATION DUNE FEUILLE "Stastistique".
# Enregistrement du résultat.
writer1.save()
```
