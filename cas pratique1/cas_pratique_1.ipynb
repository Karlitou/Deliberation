#### 10. Dans cette partie, on demandera d’écrire un programme qui permet de :
#### • créer et remplir un vecteur représentant la somme pluviométrique annuelle ;
#### • déterminer la moyenne du vecteur obtenu ;
#### • calculer la différence entre le vecteur et sa moyenne.

```python
som=np.sum(Df_pluvio.iloc[:, 2:], axis=1)
moy=np.mean(som)
ecartype=moy-som
print("Le Vecteur créé est:\n",som)
print("********************************")
print("La moyenne du Vecteur est::",moy)
print("********************************")
print("La difference entre le Vecteur et sa moyenne est:\n",ecartype)
```

#### 11. Tracer la courbe de la différence obtenue à la question précédente en fonction des années.

```python
fig6,ax6 = plt.subplots(figsize = (15,8))
plt.bar(Df_pluvio.Annees,ecartype, label = "ecatype de la pluviometrie.")
plt.title("Ecartype Pluviometrie.")
plt.grid()
plt.legend()
plt.xlabel("Annees")
plt.ylabel("Pluie")
plt.show()
```
